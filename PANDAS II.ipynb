Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')

ratings.info()
ratings
```

```python
users_100 = ratings[['userId', 'timestamp']].groupby('userId').agg(['count', 'min', 'max']).reset_index()
users_100['movieCount'] = users_100['timestamp']['count']
users_100['firstMovieTs'] = users_100['timestamp']['min']
users_100['latestMovieTs'] = users_100['timestamp']['max']
users_100 = users_100[users_100['movieCount'] > 100]
del users_100['timestamp']
```

```python
users_100['lifetime'] = users_100['latestMovieTs'] - users_100['firstMovieTs']

users_100
```

```python
users_100['lifetime'].mean()
```
